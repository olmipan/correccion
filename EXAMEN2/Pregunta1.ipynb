```python
import math

# Datos de la ecuación
a = 1
b = 62.1
c = 1

# Soluciones conocidas
x1_real = -0.016107237
x2_real = -62.083893

# Método A
discriminante = b**2 - 4*a*c
x1_A = (-b + math.sqrt(discriminante)) / (2*a)
x2_A = (-b - math.sqrt(discriminante)) / (2*a)

# Método B
x1_B = (-2*c) / (b + math.sqrt(discriminante))
x2_B = (-2*c) / (b - math.sqrt(discriminante)) if (b - math.sqrt(discriminante)) != 0 else float('nan')

# Cálculo del error relativo para cada solución
def error_relativo(x_real, x_calculado):
    return abs(x_real - x_calculado) / abs(x_real)

# Error relativo para el Método A
e1_A = error_relativo(x1_real, x1_A)
e2_A = error_relativo(x2_real, x2_A)

# Error relativo para el Método B
e1_B = error_relativo(x1_real, x1_B)
e2_B = error_relativo(x2_real, x2_B)

# Mostrar resultados
print(f"Error relativo para x1 usando Método A: {e1_A:.5e}")
print(f"Error relativo para x2 usando Método A: {e2_A:.5e}")
print(f"Error relativo para x1 usando Método B: {e1_B:.5e}")
print(f"Error relativo para x2 usando Método B: {e2_B:.5e}")
```

```python
import math

# Función para calcular las raíces usando el Método A
def metodo_A(a, b, c):
    discriminante = b**2 - 4*a*c
    if discriminante < 0:
        return None  # Si el discriminante es negativo, no hay soluciones reales
    raiz_discriminante = math.sqrt(discriminante)
    x1 = (-b + raiz_discriminante) / (2*a)
    x2 = (-b - raiz_discriminante) / (2*a)
    return x1, x2

# Función para calcular las raíces usando el Método B
def metodo_B(a, b, c):
    discriminante = b**2 - 4*a*c
    if discriminante < 0:
        return None  # Si el discriminante es negativo, no hay soluciones reales
    raiz_discriminante = math.sqrt(discriminante)
    x1 = (-2*c) / (b + raiz_discriminante)
    x2 = (-2*c) / (b - raiz_discriminante) if (b - raiz_discriminante) != 0 else float('nan')
    return x1, x2

# Función para calcular el error relativo
def error_relativo(x_real, x_calculado):
    if x_real == 0:
        return float('nan')
    return abs(x_real - x_calculado) / abs(x_real)

# Definir los coeficientes de la ecuación
a = 1
b = 62.1
c = 1

# Soluciones reales
x1_real = -0.016107237
x2_real = -62.083893

# Calcular las raíces con ambos métodos
raices_A = metodo_A(a, b, c)
raices_B = metodo_B(a, b, c)

# Calcular el error relativo para ambos métodos
if raices_A:
    e1_A = error_relativo(x1_real, raices_A[0])
    e2_A = error_relativo(x2_real, raices_A[1])
else:
    e1_A = e2_A = float('nan')

if raices_B:
    e1_B = error_relativo(x1_real, raices_B[0])
    e2_B = error_relativo(x2_real, raices_B[1])
else:
    e1_B = e2_B = float('nan')

# Mostrar los resultados
print(f"Método A: Raíz 1 = {raices_A[0] if raices_A else 'NaN'}, Raíz 2 = {raices_A[1] if raices_A else 'NaN'}")
print(f"Método B: Raíz 1 = {raices_B[0] if raices_B else 'NaN'}, Raíz 2 = {raices_B[1] if raices_B else 'NaN'}")

# Mostrar los errores relativos
print(f"Error relativo para x1 usando Método A: {e1_A:.5e}")
print(f"Error relativo para x2 usando Método A: {e2_A:.5e}")
print(f"Error relativo para x1 usando Método B: {e1_B:.5e}")
print(f"Error relativo para x2 usando Método B: {e2_B:.5e}")

# Elegir el mejor método basado en el discriminante
discriminante = b**2 - 4*a*c
if discriminante < 0:
    print("No hay soluciones reales.")
elif abs(raices_A[0] - raices_A[1]) < abs(raices_B[0] - raices_B[1]):
    print("Método A es el mejor método debido a la estabilidad numérica.")
else:
    print("Método B es el mejor método debido a la estabilidad numérica.")
```
