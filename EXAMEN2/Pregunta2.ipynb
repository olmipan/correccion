Determine todas las raíces del polinomio

$ X^5 - 6 * X^4 + 2 * X^3 + 20 * X^2 - 27 * X + 10 = 0 $

```python
from scipy.optimize import bisect
import matplotlib.pyplot as plt
import numpy as np

def funcion(x: float):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

# Definimos el rango de valores de x para el gráfico
x = np.linspace(-5, 5, 100)

# Graficamos la función
plt.plot(x , [funcion(i) for i in x])
plt.axhline(0, color='black',linewidth=1)  # Línea horizontal en y=0
plt.axvline(0, color='black',linewidth=1)  # Línea vertical en x=0
plt.title("Gráfico de la función")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()

# Raíz 1: Intervalo [-5, 0] (aseguramos cambio de signo)
root1 = bisect(funcion, a=-5, b=0)
print(f"Raíz 1: {root1}")

# Raíz 2: Intervalo [0, 2] (aseguramos cambio de signo)
root2 = bisect(funcion, a=0, b=2)
print(f"Raíz 2: {root2}")

# Raíz 3: Intervalo [2, 5] (aseguramos cambio de signo)
root3 = bisect(funcion, a=2, b=5)
print(f"Raíz 3: {root3}")
```
