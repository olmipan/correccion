```python
import matplotlib.pyplot as plt
import numpy as np

from src import (
    eliminacion_gaussiana,
    descomposicion_LU,
    resolver_LU,
    matriz_aumentada,
    separar_m_aumentada,
)

def eliminacion_gaussiana(A: np.ndarray) -> np.ndarray:
    """Resuelve un sistema de ecuaciones lineales mediante el método de eliminación gaussiana.

    ## Parameters

    ``A``: matriz aumentada del sistema de ecuaciones lineales. Debe ser de tamaño n-by-(n+1), donde n es el número de incógnitas.

    ## Return

    ``solucion``: vector con la solución del sistema de ecuaciones lineales.

    """
    if not isinstance(A, np.ndarray):
        print("Convirtiendo A a numpy array.")
        A = np.array(A, dtype=float)
    assert A.shape[0] == A.shape[1] , "La matriz A debe ser de tamaño n-by-n."
    n = A.shape[0]
    numcambios = 0
    for i in range(0, n - 1):  # loop por columna

        # --- encontrar pivote
        p = None  # default, first element
        for pi in range(i, n):
            if A[pi, i] == 0:
                # must be nonzero
                continue

            if p is None:
                # first nonzero element
                p = pi
                continue

            if abs(A[pi, i]) < abs(A[p, i]):
                p = pi

        if p is None:
            # no pivot found.
            raise ValueError("No existe solución única.")

        if p != i:
            # swap rows
            print(f"Intercambiando filas {i} y {p}")
            _aux = A[i, :].copy()
            A[i, :] = A[p, :].copy()
            A[p, :] = _aux
            numcambios+=1

        # --- Eliminación: loop por fila
        for j in range(i + 1, n):
            m = A[j, i] / A[i, i]
            A[j, i:] = A[j, i:] - m * A[i, i:]

        print(f"\n{A}")

    
    return A,numcambios


# ####################################################################
def calc_determinante(A: list[list[float]]) -> float:
    """Función que calcula el determinante usando el método...
    [Descomposición LU, eliminación gaussiana, Gauss-Jordan, Gauss-Jacobi o Gauss-Seidel]

    ## Parameters
    ``A``: Matriz cuadrada de tamaño n x n

    ## Return
    ``detA``: Determinante de la matriz A

    """
    print("Matriz escalonada: \n\n")
    Det=1
    B,cambios=eliminacion_gaussiana(A)
    for fila in B:
        print(fila)

    for i in range(len(B)):
        for j in range(len(B)): 
            if i==j: 
                Det*=B[i,j]
                print(Det)
        print("\n")        
            
    # completar
    return  Det*(-1)**cambios
```

```python
from src import ajustar_min_cuadrados
#Derivadas parcales para la funcion cuadratica
def der_parcial_0x2(xs: list, ys: list) -> tuple[float, float, float,float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 0 al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
    c_2 * a_2 ac_1 * a_1 + c_0 * a_0 = c_ind
    ## Parameters

    ``xs``: lista de valores de x.

    ``ys``: lista de valores de y.

    ## Return
    
    ``c_2``: coeficiente del parámetro 2.

    ``c_1``: coeficiente del parámetro 1.

    ``c_0``: coeficiente del parámetro 0.

    ``c_ind``: coeficiente del término independiente.

    """
    c_2 = 0
    c_1 = 0
    c_0 = 0
    c_ind = 0

    # coeficiente del término independiente
    c_ind = sum(ys)

    # coeficiente del parámetro 2
    c_2 = sum([x**2 for x in xs])

    # coeficiente del parámetro 1
    c_1 = sum(xs)

    # coeficiente del parámetro 0
    c_0 = len(xs)
    
    return (c_2, c_1, c_0, c_ind)

def der_parcial_1x2(xs: list, ys: list) -> tuple[float, float, float,float]:  
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 1 al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
    c_2 * a_2 ac_1 * a_1 + c_0 * a_0 = c_ind
    ## Parameters
    ``xs``: lista de valores de x.
    ``ys``: lista de valores de y.
    ## Return
    ``c_2``: coeficiente del parámetro 2.
    ``c_1``: coeficiente del parámetro 1.
    ``c_0``: coeficiente del parámetro 0.
    ``c_ind``: coeficiente del término independiente.
    """
    c_2 = 0
    c_1 = 0
    c_0 = 0
    c_ind = 0
    # coeficiente del término independiente
    c_ind = sum([x*y for x,y in zip(xs,ys)])
    # coeficiente del parámetro 2
    c_2 = sum([x**3 for x in xs])
    # coeficiente del parámetro 1
    c_1 = sum([x**2 for x in xs])
    # coeficiente del parámetro 0
    c_0 = sum(xs)
    return (c_2, c_1, c_0, c_ind) 

def der_parcial_2x2(xs: list, ys: list) -> tuple[float, float, float,float]:  
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 1 al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
    c_2 * a_2 ac_1 * a_1 + c_0 * a_0 = c_ind
    ## Parameters
    ``xs``: lista de valores de x.
    ``ys``: lista de valores de y.
    ## Return
    ``c_2``: coeficiente del parámetro 2.
    ``c_1``: coeficiente del parámetro 1.
    ``c_0``: coeficiente del parámetro 0.
    ``c_ind``: coeficiente del término independiente.
    """
    c_2 = 0
    c_1 = 0
    c_0 = 0
    c_ind = 0
    # coeficiente del término independiente
    c_ind = sum([x**2*y for x,y in zip(xs,ys)])
    # coeficiente del parámetro 2
    c_2 = sum([x**4 for x in xs])
    # coeficiente del parámetro 1
    c_1 = sum([x**3 for x in xs])
    # coeficiente del parámetro 0
    c_0 = sum([x**2 for x in xs])
    return (c_2, c_1, c_0, c_ind)

xs1 = [
    -5.0000,
    -3.8889,
    -2.7778,
    -1.6667,
    -0.5556,
    0.5556,
    1.6667,
    2.7778,
    3.8889,
    5.0000,
]
ys1 = [
    57.2441,
    33.0303,
    16.4817,
    7.0299,
    0.5498,
    0.7117,
    3.4185,
    12.1767,
    24.9167,
    44.2495,
]

a2,a1,a0 = ajustar_min_cuadrados(xs1, ys1, gradiente=[der_parcial_0x2, der_parcial_1x2, der_parcial_2x2])

def x2(x):
    return a2*x**2 + a1*x + a0

x = np.linspace(-5, 5, 100)
y = x2(x)
plt.scatter(xs1, ys1)
plt.plot(x, y, color="red", label=r"$ y = a_1 x + a_0 $")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Conjunto de datos 1")
plt.show()
print(a2,a1,a0)
print("f(2.25)",x2(2.25))
print("f(-2.25)",x2(-2.25))
```

```python
# Derivadas parciales para exponencial
# ####################################################################
def der_parcial_1exp(xs: list, ys: list) -> tuple[float, float, float]:
    """
    Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 1 (b) al reemplazar los valores ``xs`` y ``ys``.
    La ecuación es de la forma: c_1 * A + c_0 * b = c_ind

    ## Parameters
    ``xs``: lista de valores de x.
    ``ys``: lista de valores de y.

    ## Return
    ``c_1``: coeficiente del parámetro 0 (A).
    ``c_0``: coeficiente del parámetro 1 (b).
    ``c_ind``: coeficiente del término independiente.
    """
    Y = np.log(ys)
    c_1 = sum(xs)
    c_0 = sum(x * x for x in xs)
    c_ind = sum(x * y for x, y in zip(xs, Y))
    return (c_1, c_0, c_ind)



def der_parcial_0exp(xs: list, ys: list) -> tuple[float, float, float]: 
    """
    Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 0 (A) al reemplazar los valores ``xs`` y ``ys``.
    La ecuación es de la forma: c_1 * A + c_0 * b = c_ind

    ## Parameters
    ``xs``: lista de valores de x.
    ``ys``: lista de valores de y.

    ## Return
    ``c_1``: coeficiente del parámetro 0 (A).
    ``c_0``: coeficiente del parámetro 1 (b).
    ``c_ind``: coeficiente del término independiente.
    """
    Y = np.log(ys)
    c_1 = len(xs)  # n
    c_0 = sum(xs)
    c_ind = sum(Y)
    return (c_1, c_0, c_ind)


xs2 = [
    0.0003,
    0.0822,
    0.2770,
    0.4212,
    0.4403,
    0.5588,
    0.5943,
    0.6134,
    0.9070,
    1.0367,
    1.1903,
    1.2511,
    1.2519,
    1.2576,
    1.6165,
    1.6761,
    2.0114,
    2.0557,
    2.1610,
    2.6344,
]
ys2 = [
    1.1017,
    1.5021,
    0.3844,
    1.3251,
    1.7206,
    1.9453,
    0.3894,
    0.3328,
    1.2887,
    3.1239,
    2.1778,
    3.1078,
    4.1856,
    3.3640,
    6.0330,
    5.8088,
    10.5890,
    11.5865,
    11.8221,
    26.5077,
]

a,b = ajustar_min_cuadrados(xs2, ys2, gradiente=[der_parcial_0exp, der_parcial_1exp])
 
def exp(x):
    return np.exp(a)*np.exp(b*x)
x = np.linspace(0, 3, 100)
y = exp(x)
plt.plot(x, y, color="red", label=r"$ y = A e^{bx} $")
plt.scatter(xs2, ys2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Conjunto de datos 2")
plt.show() 
print(a,b)
print("f(5)",exp(2.25))
print("f(1)",exp(-2.25))
```
